# [dbo].[v_DonHangChiTiet]

```sql
/****** Object:  View [dbo].[v_DonHangChiTiet]    Script Date: 5/15/2026 11:52:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 
-- =====================================================
-- VIEWS HỮU ÍCH
-- =====================================================
 
-- View: Đơn hàng đầy đủ thông tin
CREATE   VIEW [dbo].[v_DonHangChiTiet]
AS
SELECT 
    dh.MaDonHang,
    dh.NgayDat,
    kh.HoVaTen      AS TenKhachHang,
    kh.SoDienThoai,
    nv.HoVaTen      AS TenNhanVien,
    dh.TongTien,
    dh.TrangThai,
    dh.GhiChu
FROM DonHang dh
LEFT JOIN KhachHang kh ON dh.MaKH = kh.MaKH
LEFT JOIN NhanVien  nv ON dh.MaThanhVien = nv.MaThanhVien
GO
```

# [dbo].[sp_DoanhThuTheoThang]

```sql
/****** Object:  StoredProcedure [dbo].[sp_DoanhThuTheoThang]    Script Date: 5/15/2026 11:52:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 
-- SP: Thống kê doanh thu theo tháng
CREATE   PROCEDURE [dbo].[sp_DoanhThuTheoThang]
    @Nam INT = NULL
AS
BEGIN
    SET @Nam = ISNULL(@Nam, YEAR(GETDATE()))
    SELECT 
        MONTH(NgayDat)  AS Thang,
        COUNT(*)        AS SoDonHang,
        SUM(TongTien)   AS DoanhThu
    FROM DonHang
    WHERE YEAR(NgayDat) = @Nam AND TrangThai = N'Hoàn thành'
    GROUP BY MONTH(NgayDat)
    ORDER BY Thang
END
GO
```

# [dbo].[sp_GetSanPhamConHang]

```sql
/****** Object:  StoredProcedure [dbo].[sp_GetSanPhamConHang]    Script Date: 5/15/2026 11:52:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 
-- SP: Lấy danh sách sản phẩm còn hàng
CREATE   PROCEDURE [dbo].[sp_GetSanPhamConHang]
AS
BEGIN
    SELECT MaSanPham, TenSanPham, KieuDang, GiaXuat, 
           TinhTrang, ChatLieu, HangSanXuat,
           (SoLuongNhap - DaBan) AS TonKho
    FROM SanPham
    WHERE TinhTrang = N'Còn hàng'
    ORDER BY TenSanPham
END
GO
```

# [dbo].[sp_ThemChiTietDH]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ThemChiTietDH]    Script Date: 5/15/2026 11:52:23 AM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
 
-- SP: Thêm chi tiết đơn hàng
CREATE   PROCEDURE [dbo].[sp_ThemChiTietDH]
    @MaDonHang NVARCHAR(10),
    @MaSanPham NVARCHAR(10),
    @SoLuong   INT
AS
BEGIN
    DECLARE @DonGia INT
    SELECT @DonGia = GiaXuat FROM SanPham WHERE MaSanPham = @MaSanPham
 
    INSERT INTO ChiTietDonHang (MaDonHang, MaSanPham, SoLuong, DonGia)
    VALUES (@MaDonHang, @MaSanPham, @SoLuong, @DonGia)
 
    -- Cập nhật tổng tiền đơn hàng
    UPDATE DonHang
    SET TongTien = (
        SELECT SUM(SoLuong * DonGia) FROM ChiTietDonHang WHERE MaDonHang = @MaDonHang
    )
    WHERE MaDonHang = @MaDonHang
END
GO
```
